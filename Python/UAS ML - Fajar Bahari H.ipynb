# Model Per Kolom (Logistic Regression)

```python
models = {}
for col in df.columns[1:]:
    X = df['Pass1'].values.reshape(-1, 1)
    y = df['Pertandingan'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    model = LogisticRegression(max_iter=5000).fit(X_train, y_train)
    models[col] = model
    print("Accuracy for", col, ":", model.score(X_test, y_test))
```

# 2 Klasifikasi

```python
X = df.iloc[:, 1:].values
y = df['Pertandingan'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
model = LogisticRegression(max_iter=10000).fit(X_train, y_train)
y_pred = model.predict(X_test)
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

# 2 Clustering

```python
X = df.iloc[:, 1:].values
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print("\nCluster Centers:")
print(kmeans.cluster_centers_)
```

```python
#Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Define features
X = df.iloc[:, :-1].values

# Define number of clusters
k = 3

# Create AgglomerativeClustering model
agg_clustering = AgglomerativeClustering(n_clusters=k)

# Fit model to data
y_pred = agg_clustering.fit_predict(X)

# Plot scatter plot with different colors for each cluster
colors = ['red', 'blue', 'green']
for i in range(k):
    plt.scatter(X[y_pred == i, 0], X[y_pred == i, 1], s=30, c=colors[i], label='Cluster ' + str(i+1))
plt.legend()
plt.show()
```
